## 4. Conditionals

**Example 1:** If we want to identify if a given number is odd or even then the control flow is as follows:

```python
n = int(input())

if n % 2 == 0:
    print("n is even")
else:
    print("n is odd")
```

**Example 2:** Identify if a given number if positive, negative or 0

```python
n = float(input())

if n > 0:
    print("n is positive")
elif n < 0:
    print("n is negative")
else:
    print("n is 0")
```

## 5. While Loop

The syntax of a while loop is as follows:

while *condition* :

    statements

The while loop will run as long as the *condition* is satisfied, therefore it is imperative to make sure the *condition* is not satisfied at some point in real time. If the breaking condition is not met then the code will never end.

**Example :** Print the first 10 even numbers

```python
i = 1
while i <= 10:
    print(2*i)
    i += 1
```
